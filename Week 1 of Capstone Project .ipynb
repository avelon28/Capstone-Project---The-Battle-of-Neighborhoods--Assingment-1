## Providing temporary accodomation for healthcare workers in London

## Introduction

### London is one the cities that has being hit severly with the COVID-19 pandemic, in which the number of cases and death rose astronomically. This means that the National Health Service (NHS) is overwhelmed with patients, not enough staff, hence causing a risk for the family or friends who live with the healthcare workers. Therefore, the idea is to find a location for temporary accomodation nearby hospitals in London for healthcare workers when they are COVID positive, so they can minimize the risk of exposing COVID to their family or friends who live with them. Therefore, the target audience for this project is the healthcare workers, and patients who are positive for COVID. 

## Business Problem

### The objective is to analyse and select the hospitals who receive the most requests for COVID, together with the number of healthcare workers who are positive in the busiest hospitals in London. Using methods of data science and machine learning like clustering, is intended to find the solution to this idea.
